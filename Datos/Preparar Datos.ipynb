```python
# tweet_df = pd.read_csv('Twint_Tweets_2020-11-01_2020-12-01.csv', index_col=0)
tweet_df = pd.read_csv('Twint_Tweets_2019-10-01_2020-11-01.csv', index_col=0)
# tweet_df = pd.read_csv('Twint_Tweets_2019-10-01_2020-12-01.csv', index_col=0)
tweet_df = tweet_df.drop(["timezone", "day", "hour", "link", "user_id", "user_id_str", "name", "urls", "photos", "video", "thumbnail", "quote_url", "search", "geo", "near", "source", "translate", "trans_src", "trans_dest"], axis=1)
tweet_df = tweet_df[55000:]
tweet_df
```

```python
clean_tweets = []
clean_tweets.clear()
lemma_tweets = []
lemma_tweets.clear()
i = 0
nlp = spacy.load("es_core_news_sm")
lemmatizer = SpacyCustomLemmatizer()
nlp.add_pipe(lemmatizer, name="lemmatizer", after="tagger")
tokenizer = RegexpTokenizer(r'\w+')
for tweet in tweet_df['tweet']:
    tweet = re.sub(r'https?:\/\/\S*', '', tweet, flags=re.MULTILINE)
    tweet = re.sub(r'@[A-Za-z0-9]+', '', tweet, flags=re.MULTILINE)
    tweet = tweet.lower()
    lemma_tokens = []
    lemma_tokens.clear()
    clean_tokens = []
    clean_tokens.clear()
    for token in tokenizer.tokenize(tweet):
        if token not in stopwords.words('spanish'):
            lemma_tokens.append(nlp(token)[0].lemma_)
            clean_tokens.append(token)
    lemma_tweets.append(lemma_tokens)
    clean_tweets.append(clean_tokens)
    if i % 100 == 0:
        print(i)
    i += 1
tweet_df['clean_tweet'] = clean_tweets
clean_tweets.clear()
tweet_df['lemma_tweet'] = lemma_tweets
lemma_tweets.clear()
tweet_df.to_csv('Clean_Tweets_05_9.csv', index=False, header=True)
```
